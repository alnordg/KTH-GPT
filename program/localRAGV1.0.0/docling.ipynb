```python
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.document_converter import DocumentConverter, PdfFormatOption

converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=PdfPipelineOptions(do_picture_description=False)),
        InputFormat.DOCX: PdfFormatOption(pipeline_options=PdfPipelineOptions()),
        InputFormat.PPTX: PdfFormatOption(pipeline_options=PdfPipelineOptions()),
        InputFormat.MD: PdfFormatOption(pipeline_options=PdfPipelineOptions()),
    }
)

source = "test.pdf"
converter = DocumentConverter()
doc = converter.convert(source).document
```

```python
print(doc.export_to_markdown()[:2000])
```
